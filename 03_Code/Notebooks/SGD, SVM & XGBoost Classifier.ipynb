```python
train, val = Utl.split_dataset(df, y_col="class", test_size=0.2, with_stratify=True, shuffle=True)
#train, test = Utl.split_dataset(train, y_col="class", test_size=0.1, with_stratify=True, shuffle=True)


X_train = train["content"].values
y_train = train["class"].values
X_val = val["content"].values
y_val = val["class"].values
```

```python
num_classes = Utl.get_nb_classes(y_train)
label_encoder = Utl.get_label_encoder_obj(y_train)
y_train = Utl.get_y_label_encoder(label_encoder, y_train)
y_val = Utl.get_y_label_encoder(label_encoder, y_val)
```

```python
path_directory = "../Weights"
file_name = "label_encoder.pkl"
Utl.save_model_pkl(label_encoder, path_directory, file_name)
file_name = "../Weights/label_encoder.pkl"
label_encoder = Utl.load_model_pkl(file_name)
```

```python
X_train_numeric = Utl.log_transform(train[numeric_cols].values)
X_val_numeric = Utl.log_transform(val[numeric_cols].values)

minmax_scaler = Utl.minmax_scaler_fit(X_train_numeric)
X_train_numeric = Utl.minmax_scaler_transform(minmax_scaler, X_train_numeric)
X_val_numeric = Utl.minmax_scaler_transform(minmax_scaler, X_val_numeric)

path_directory = "../Weights"
file_name = "minmax_scaler.pkl"
Utl.save_model_pkl(minmax_scaler, path_directory, file_name)
file_name = "../Weights/minmax_scaler.pkl"
minmax_scaler = Utl.load_model_pkl(file_name)
```

```python
cat_fet_cols_name = ["content_name", "campaign_name"]
cat_fet_cols_name
```

```python
cat_fet_encode_dict, X_train_ohe = Utl.ohe_features_fit(train, cat_fet_cols_name)

feat_directory = "../Weights/FeaturesEncoder"
Utl.save_features_encoder(cat_fet_encode_dict, feat_directory)
cat_fet_encode_dict = Utl.load_features_encoder(feat_directory)

X_val_ohe = Utl.ohe_features_transform(val, cat_fet_encode_dict)
```

```python
cat_fet_encode_dict
```

```python
X_train_tfidf = Utl.horizontal_feat_concatenate_csr(X_train_ohe, X_train_numeric)
X_val_tfidf = Utl.horizontal_feat_concatenate_csr(X_val_ohe, X_val_numeric)
```

```python
class_weight = Utl.get_class_weights(y_train)
```

```python
X_train_numeric.shape
```

```python
model, _ = SGD_clf.fit_SGDClassifier(X_train_tfidf, y_train.ravel(), class_weight)
directory = "../Weights"
SGD_clf_file = "SGD_clf.pkl"
Utl.save_model_pkl(model, directory, SGD_clf_file)
SGD_clf_file = "../Weights/SGD_clf.pkl"
model = Utl.load_model_pkl(SGD_clf_file)
y_hat = model.predict(X_val_tfidf)
Utl.get_prediction_results(y_val, y_hat, label_encoder, num_classes)
y_hat = Utl.predict(model, X_val_tfidf)
```

```python
preprocessing_pipeline = Fe.fit_preprocessing_pipeline(X_train, using_tfidf_vec=False,
                                                       ngram_range=(1, 1), use_idf=True,
                                                       max_df=0.5, min_df=10, max_features=5000)
directory = "../Weights"
preprocessing_pipeline_file = "preprocessing_pipeline.pkl"
Utl.save_model_pkl(preprocessing_pipeline, directory, preprocessing_pipeline_file)
preprocessing_pipeline_file = "../Weights/preprocessing_pipeline.pkl"
preprocessing_pipeline = Utl.load_model_pkl(preprocessing_pipeline_file)
X_train_tfidf = Fe.transform_preprocessing_pipeline(preprocessing_pipeline, X_train)
X_val_tfidf = Fe.transform_preprocessing_pipeline(preprocessing_pipeline, X_val)
```

```python
using_cat_feats = True
using_numeric_feats = True
```

```python
if using_cat_feats:
    X_train_tfidf = Utl.horizontal_feat_concatenate_csr(X_train_tfidf, X_train_ohe)
    X_val_tfidf = Utl.horizontal_feat_concatenate_csr(X_val_tfidf, X_val_ohe)

if using_numeric_feats:
    X_train_tfidf = Utl.horizontal_feat_concatenate_csr(X_train_tfidf, X_train_numeric)
    X_val_tfidf = Utl.horizontal_feat_concatenate_csr(X_val_tfidf, X_val_numeric)
```

```python
model, _ = SGD_clf.fit_SGDClassifier(X_train_tfidf, y_train.ravel(), class_weight)
directory = "../Weights"
SGD_clf_file = "SGD_clf.pkl"
Utl.save_model_pkl(model, directory, SGD_clf_file)
SGD_clf_file = "../Weights/SGD_clf.pkl"
model = Utl.load_model_pkl(SGD_clf_file)
y_hat = model.predict(X_val_tfidf)
Utl.get_prediction_results(y_val, y_hat, label_encoder, num_classes)
y_hat = Utl.predict(model, X_val_tfidf)
```

```python
model, _ = SVM_clf.fit_SVMClassifier(X_train_tfidf, y_train.ravel(), None)
directory = "../Weights"
SVM_clf_file = "SVM_clf.pkl"
Utl.save_model_pkl(model, directory, SVM_clf_file)
SVM_clf_file = "../Weights/SVM_clf.pkl"
model = Utl.load_model_pkl(SVM_clf_file)
y_hat = model.predict(X_val_tfidf)
Utl.get_prediction_results(y_val, y_hat, label_encoder, num_classes)
y_hat = Utl.predict(model, X_val_tfidf)
```
